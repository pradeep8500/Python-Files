```python
def chunk_search(segment, chunked):
    m = len(chunked)
    print(chunked[0][0][0])
    list1 = []
    for j in range(m):
        print('jjj',j)
        if (len(chunked[j]) > 2 or len(chunked[j]) == 1):
            list1.append(j)
        if (len(chunked[j]) == 2):
            try:
                str1 = chunked[j][0][0] + " " + chunked[j][1][0]
            except Exception:
                pass
            else:
                if (str1 in segment) == True:
                    list1.append(j)
    return list1
import nltk
words = nltk.word_tokenize('My best friend and I have been studying in the same school since kindergarten.')
tagged = nltk.pos_tag(words) 
#chunkGram = r"""Chunk: {<RB.?>*<VB.?>*<NNP><NN>?}"""
chunkGram = r"""chunk:{<TO>+<DT>?<RB.?>*<JJ.?>*<NN.?|PRP|PRP\$|VBG|DT|POS|CD|VBN>+}"""
chunkParser = nltk.RegexpParser(chunkGram)
chunked = chunkParser.parse(tagged)
chunk_search('My best friend and I have been studying in the same school since kindergarten.',chunked)
```
